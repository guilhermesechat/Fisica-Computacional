```python
import matplotlib.pyplot as plt
import numpy as np
import gif
np.set_printoptions(threshold=np.inf)


grid_size = 71
rule = 24
time = 100
initial = 'Glidergun'

#create a grid matrix
Grid = np.zeros((grid_size+2, grid_size+2))
#a draft matrix to be used inside the loop
newGrid = np.zeros((grid_size+2, grid_size+2))

#writing the initial matrix states
if initial == 'glider':
    for y in range(3):
        Grid[int((grid_size+2)/2)-y][int((grid_size+2)/2)] = 1
    for y in range(2):
            Grid[int((grid_size+2)/2)][int((grid_size+2)/2)+y] = 1
    Grid[int((grid_size+2)/2)-1][int((grid_size+2)/2)+2] = 1
elif initial == 'blinker':
    for y in range(3):
        Grid[int((grid_size+2)/2)+y][int((grid_size+2)/2)] = 1
elif initial == "G1":
    for y in range(3):
        Grid[int((grid_size+2)/2)+y][int((grid_size+2)/2)] = 1
    for y in range(3):
        Grid[int((grid_size+2)/2) + 1][int((grid_size+2)/2)+1+y] = 1    
elif initial == "MWSS":
    for y in range(2):
        Grid[int((grid_size+2)/2)][int((grid_size+2)/2 +4+ y)] = 1
    for y in range(7):
        Grid[int((grid_size+2)/2) + 1][int((grid_size+2)/2)+y] = 1  
    Grid[int((grid_size+2)/2) + 1][int((grid_size+2)/2)+4] = 0 
    for y in range(6):
        Grid[int((grid_size+2)/2) + 2][int((grid_size+2)/2)+y] = 1    
    for y in range(4):
        Grid[int((grid_size+2)/2) + 3][int((grid_size+2)/2)+1+y] = 1   
elif initial == "Glidergun":
    gg1 = [25]
    gg2 = [23,25]
    gg3 = [13,14,21,22,35,36]
    gg4 = [12,16,21,22,35,36]
    gg5 = [1,2,11,17,21,22]
    gg6 = [1,2,11,15,17,18,23,25]
    gg7 = [11,17,25]
    gg8 = [12,16]
    gg9 = [13,14]
    gg = [gg1,gg2,gg3,gg4,gg5,gg6,gg7,gg8,gg9]    
    r=0
    for z in gg:
        for y in range(len(z)):
            Grid[int((grid_size+2)/3) + r][int((grid_size+2)/3)+ z[y]]=1
        r += 1
        

#a rule in binary number
x = '{0:b}'.format(rule).rjust(10,'0')

#dictionary that relates elements of the x string to the sum values
comb = {'9':x[0],'8':x[1],'7':x[2],'6':x[3],'5':x[4],'4':x[5],'3':x[6],'2':x[7],'1':x[8],'0':x[9]}


@gif.frame
def plotca(grid_size):
    for l in range(1,grid_size+1):
        for m in range(1,grid_size+1):
    #it's a totalistic cellular automaton, so the nearest neighborhoods should be added to relate to the dictionary
            a = str(int(Grid[m][l]+Grid[m-1][l]+Grid[m+1][l]+Grid[m][l-1]+Grid[m][l+1]+Grid[m+1][l+1]+Grid[m-1][l-1]+Grid[m+1][l-1]+Grid[m-1][l+1]))
            if Grid[m][l] ==1:
                newGrid[m][l] = comb[a]
            elif a==str(3):
                newGrid[m][l] = 1
    #here the draft matrix is made equal to the stanard on the be used as new in the new time loop
    Grid[:,:] = newGrid[:,:]
    fig, ax = plt.subplots(1,1,figsize=(10, 10))
    ax.imshow(Grid, interpolation='nearest')
    #fig.savefig('Nova/' + str(i) + '.png')

frames = []
for i in range(time): 
    frame = plotca(grid_size)
    frames.append(frame)

gif.save(frames, initial + ".gif", duration = 100)    
```
