```python
# things to import
import itertools  
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
np.set_printoptions(threshold=np.inf)
#Carsten Hensel
```

- We want to build a simple NN:
    - (2+1) inputs
    - 1 hidden neuron
    - one output 

In Keras this can be done with a `Sequential` model.

```python
size = 10
```

```python
#data
inputt = np.array([i for i in itertools.product([0,1], repeat = size)])
np.random.shuffle(inputt)
len(inputt)
```

```python
model = Sequential()
```

This represents a stack of layers. Let's add a layer. For our example we need a hidden layer with one output, two inputs and a bias term:

```python
model.add(Dense(size, activation='sigmoid', use_bias=True, input_dim=size))
model.add(Dense(35*size, activation='sigmoid', use_bias=True))
model.add(Dense(30*size, activation='sigmoid', use_bias=True))
model.add(Dense(35*size, activation='sigmoid', use_bias=True))
model.add(Dense(35*size, activation='sigmoid', use_bias=True))
model.add(Dense(size, activation='sigmoid', use_bias=True))
```

Before training you need to 'compile' the Sequence. The compilation configures the learning process. In our case we want to configure our stack for a binary classification problem:

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
train_data = int((2**size)/3)
```

```python
batch = 32
what = model.fit(inputt[:train_data], B[:train_data], validation_data=(inputt[train_data:], B[train_data:]), epochs=1200, batch_size=batch)
```

- x_train contains the input data
- y_train contains the output data
- The batch size is a number of samples processed before the model is updated. The number of epochs is the number of complete passes through the training dataset. The size of a batch must be more than or equal to one and less than or equal to the number of samples in the training dataset

```python
plt.title("Loss")
plt.plot(range(len(what.history['loss'])), what.history['loss'], '.', label="Train");
plt.plot(range(len(what.history['val_loss'])), what.history['val_loss'], '.',label="Validation");
plt.legend();
```

```python
plt.title("Accuracy")
plt.plot(range(len(what.history['accuracy'])), what.history['accuracy'], '.', label="Train");
plt.plot(range(len(what.history['val_accuracy'])), what.history['val_accuracy'], '.',label="Validation");
plt.legend();
```

```python
model.evaluate(inputt[train_data:], B[train_data:], batch_size=1024)
```
